```python
from google.colab.patches import cv2_imshow
import cv2

img = cv2.imread("/content/astronot.jpg")
cv2_imshow(img)

img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

```python
img_blur = cv2.GaussianBlur(img, (15, 15), 0) #Bulanıklaştırma
cv2_imshow(img_blur)
```

```python
edges = cv2.Canny(img_gray, 100, 200) # Kenar tespiti
cv2_imshow(edges)
```

```python
(h, w) = img.shape[:2] # Görüntüyü döndürme
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, 45, 1.0)  # 45 derece döndürme
rotated = cv2.warpAffine(img, M, (w, h))
cv2_imshow(rotated)
```
