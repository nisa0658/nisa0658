```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
from nltk.stem import WordNetLemmatizer

def ozetleme(metin):
    sentences = sent_tokenize(metin, language='turkish')
    stop_words = set(stopwords.words('turkish'))
    lemmatizer = WordNetLemmatizer()

    words = word_tokenize(metin, language='turkish')
    words = [word.lower() for word in words if word.isalpha()]
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    freq_dist = FreqDist(words)
    sentence_scores = {}

    for i, sentence in enumerate(sentences):
        sentence_words = word_tokenize(sentence, language='turkish')
        sentence_words = [word.lower() for word in sentence_words if word.isalpha()]
        sentence_score = sum(freq_dist[word] for word in sentence_words if word in freq_dist)
        sentence_scores[i] = sentence_score

    sorted_sentences = sorted(sentence_scores.items(), key=lambda item: item[1], reverse=True)
    ozet_cumleler = [sentences[i] for i, score in sorted_sentences[:3]]  # İlk 3 cümleyi alıyoruz

    return ' '.join(ozet_cumleler)

# Metni özetlemek
metin = """
Dünya’nın Yeni Uydusu: Mini Ay
Cansu Aydın
25/09/2024

“Mini ay” olarak görülen, yaklaşık 10 metre çapındaki bir asteroidin 29 Eylül'de Dünya'nın yörüngesine girmesi ve iki aya yakın bir süre boyunca Dünya’ya eşlik etmesi bekleniyor.

Dünya’nın Yeni Uydusu: Mini Ay
dottedhippo/iStockphoto.com

Dünya’ya yaklaşan bir asteroit görüntüsü

Güneş etrafında, Mars ve Jüpiter arasında belirli bir yörüngede dolanan ve farklı boyutları ile biçimleri olan kaya ve metal parçalarına asteroit denir.

asteroit

Adastra/gettyimages.com

Farklı boyutlardaki asteroitler

Yakın zamanda keşfedilen, yaklaşık 10 metre çaplı bu asteroit 2024 PT5 olarak adlandırılıyor. “Dünya'nın geçici mini ayı”  olarak nitelenen asteroidin 29 Eylül 2024 tarihinde Dünya yörüngesine girerek gezegenimizin etrafında yaklaşık iki ay boyunca dolanması bekleniyor. 2024 PT5’in 25 Kasım'da Dünya'nın çekiminden kurtulup 9 Ocak 2025'te gezegenimize tekrar yaklaşacağı ve bu sefer yaklaşık 1,8 milyon kilometre uzaktan geçeceği tahmin ediliyor.

2024 PT5, daha önce 1960’ta, 1982’de ve 2003’te Dünya'nın yakınından geçmişti. Asteroit, üç geçişinde de Dünya’ya yaklaşık 8,2 milyon kilometre mesafeye kadar yaklaşmıştı. Normal şartlarda asteroitler Dünya'nın yerçekimi etkisi ile atmosfere girerek yanar ya da yerçekiminden etkilense de atmosfere girmeden yollarına devam eder. Bu nedenle, yaklaşık 10 metre çapındaki bu asteroidin 29 Eylül'de Dünya'nın yörüngesine girmesi ve iki aya yakın bir süre boyunca Dünya'nın yörüngesinde dolanması bilim insanlarında büyük bir merak uyandırdı.

2024 PT5’in 2055 yılında bir kez daha Dünya’ya yaklaşacağı fakat bu sefer geçişinin 5,2 milyon kilometre mesafeden gerçekleşeceği öngörülüyor.
"""

ozet = ozetleme(metin)
print(ozet)
```
